```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
C1 = np.array([1,1])
C2 = np.array([15,15])
C3 = np.array([30,30])

D1 = np.random.randn(200, 2) + C1
D2 = np.random.randn(200,2) + C2
D3 = np.random.randn(200,2) + C3

X = np.concatenate((D1, D2, D3), axis = 0)

plt.scatter(X[:,0], X[:,1], s=10)
```

```python
def repaint(X,centroids):
    plt.scatter(x=X[:,0], y=X[:,1], s=20)
    plt.scatter(x=centroids[:,0], y=centroids[:,1], s=200, marker='P')
    plt.show()
```

```python
def get_centroids(X, n_clusters):
    m , dim = X.shape
    centroids = np.zeros((n_clusters, dim))
    random_choice = np.random.randint(0 , m , n_clusters)
    for idx, value in enumerate(random_choice):
        centroids[idx] = X[value]
    return centroids
```

```python
def k_means(X , n_clusters, n_iter = 4, error = 1e-4):
    m , _ = X.shape
    centroids = get_centroids(X, n_clusters)
    clusters = [[] for _ in range(n_clusters)]

    for i in range(n_iter):
        
        repaint(X, centroids)

        for example in X:
            example  = example.reshape(1,-1)
            
            distance = np.sqrt(np.sum( (centroids - example)**2 , axis=1))
            
            clusters[np.argmin(distance)].append(example)
            
    
        
        for idx, cluster in enumerate(clusters):
        
            centroids[idx] = np.mean(np.array(cluster), axis = 0)
        
        
        clusters = [[] for _ in range(n_clusters)]
        
```

```python
k_means(X, n_clusters=3, n_iter=5)
```
